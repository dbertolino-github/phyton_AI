```python
# generate 2d classification dataset
X, y = make_circles(n_samples=100, noise=0.05)

data_0_x = []
data_0_y = []
data_1_x = []
data_1_y = []

for index, label in enumerate(y):
    features = X[index, :]
    if(label == 0): 
        data_0_x.append(features[0]) 
        data_0_y.append(features[1]) 
    else: 
        data_1_x.append(features[0]) 
        data_1_y.append(features[1]) 
        
colors = ['red', 'blue']
plt.scatter(data_0_x, data_0_y, alpha=0.8, c=colors[0], label="Class 0")
plt.scatter(data_1_x, data_1_y, alpha=0.8, c=colors[1], label="Class 1")
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.legend();


    
```
